```python
import matplotlib.pyplot as plt
import numpy as np


c_num_list = []
enc_list = []
pt_list = []
dt_list = []
pt_log = 'output/log.txt'
```

```python
with open(pt_log, 'r') as f1:
    for line in f1.readlines():
        if line[0] != 'C':
            l = line.split('\t')[:-1]
            print(l)
            c_num_list.append(int(l[0].split('.')[0]))
            enc_list.append(float(l[1]))
            pt_list.append(float(l[2]))

# with open(hg_log, 'r') as f2:
#     for line in f2.readlines():
#         if line[0] != 'E':
#             hg_val_list.append(float(line.split('\t')[-2]))
c_num_list
```

```python
x = np.array(c_num_list)
y_pt = np.array(pt_list)
y_enc = np.array(enc_list)
y_dt = np.array(dt_list)

plt.xlabel('Ciphertext Num')
plt.ylabel('Time')

plt.plot(x, y_pt, color='red', label='pTest Time')
plt.plot(x, y_enc, color='skyblue', label='Encrypt Time')

plt.rcParams['figure.dpi'] = 200
plt.legend()
plt.show()
```
