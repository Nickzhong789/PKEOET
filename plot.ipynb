```python
import matplotlib.pyplot as plt
import numpy as np


c_num_list = []
pkeoet_enc_list = []
pkeoet_ptest_list = []
pkeoet_total_list = []
pkeet_enc_list = []
dt_list = []
pkeet_t_list = []

pkeoet_log = 'output/log1.txt'
dt_log = 'output/log_d.txt'
pkeet_log = 'output/pkeet_log.txt'
```

```python
c_num_list = []
pkeoet_enc_list = []
pkeoet_ptest_list = []
pkeoet_total_list = []

with open(pkeoet_log, 'r') as f1:
    for line in f1.readlines():
        if line[0] != 'C':
            l = line.split('\t')[:-1]
            c_num_list.append(int(l[0].split('.')[0]))
            pkeoet_enc_list.append(float(l[1]))
            pkeoet_ptest_list.append(float(l[2]))
            pkeoet_total_list.append(float(l[2]))

len(c_num_list)
```

```python
dt_list = []
i = 0
with open(dt_log, 'r') as f2:
    for line in f2.readlines():
        if line[0] != 'V':
            l = line.split('\t')[:-1]
            num = int(l[0].split('.')[0])
            if num in c_num_list:
                pkeoet_total_list[i] += float(l[1])
                dt_list.append(float(l[1]))
                i += 1
len(dt_list)
```

```python
# c_num_list = []
pkeet_enc_list = []
pkeet_t_list = []

with open(pkeet_log, 'r') as f3:
    for line in f3.readlines():
        if line[0] != 'C':
            l = line.split('\t')[:-1]
            num = int(l[0].split('.')[0])
            if num in c_num_list:
                pkeet_enc_list.append(float(l[1]))
                pkeet_t_list.append(float(l[2]))

len(pkeet_t_list)
```

```python
x = np.array(c_num_list)
y_pkeoet_enc = np.array(pkeoet_enc_list)
y_pkeoet_pt = np.array(pkeoet_ptest_list)
y_pkeoet_total = np.array(pkeoet_total_list)
y_dt = np.array(dt_list)
y_pkeet_enc = np.array(pkeet_enc_list)
y_pkeet_t = np.array(pkeet_t_list)

plt.xlabel('Ciphertext Num')
plt.ylabel('Time')

# plt.plot(x, y_pkeoet_enc, color='green', label='Pkeoet Enc Time')
plt.plot(x, y_pkeoet_pt, color='green', label='Pkeoet PTest Time')
plt.plot(x, y_pkeoet_total, color='red', label='Pkeoet Total Time')
plt.plot(x, y_dt, color='orange', label='Pkeoet DTest Time')
# plt.plot(x, y_pkeet_enc, color='red', label='Pkeet Enc Time')
plt.plot(x, y_pkeet_t, color='skyblue', label='Pkeet Test Time')

plt.rcParams['figure.dpi'] = 200
plt.legend()
plt.show()
```
